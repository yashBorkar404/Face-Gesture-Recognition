```python
# Install required modules
!pip install tensorflow sklearn matplotlib

import tensorflow as tf
from tensorflow.keras.applications import MobileNet, ResNet50
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, optimizers
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
import matplotlib.pyplot as plt
import os
import shutil

# Paths
original_data_dir = "/kaggle/input/ck-dataset/CK+_dataset"
base_dir = "organized_ckplus_dataset"

# Create base directory
if not os.path.exists(base_dir):
    os.mkdir(base_dir)

# Define train, val, and test directories
train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'val')
test_dir = os.path.join(base_dir, 'test')

for dir_path in [train_dir, val_dir, test_dir]:
    if not os.path.exists(dir_path):
        os.mkdir(dir_path)

# Split data manually
classes = os.listdir(original_data_dir)
for class_name in classes:
    class_path = os.path.join(original_data_dir, class_name)
    images = os.listdir(class_path)
    train_images, test_images = train_test_split(images, test_size=0.2, random_state=42)
    val_images, test_images = train_test_split(test_images, test_size=0.5, random_state=42)

    for dataset, dataset_dir in zip([train_images, val_images, test_images], [train_dir, val_dir, test_dir]):
        class_dataset_dir = os.path.join(dataset_dir, class_name)
        if not os.path.exists(class_dataset_dir):
            os.mkdir(class_dataset_dir)

        for image in dataset:
            src = os.path.join(class_path, image)
            dst = os.path.join(class_dataset_dir, image)
            shutil.copyfile(src, dst)

# Image dimensions
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32

# Data preprocessing
datagen = ImageDataGenerator(rescale=1.0/255.0)

train_data = datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='sparse'  # Sparse categorical labels
)

val_data = datagen.flow_from_directory(
    val_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='sparse'
)

test_data = datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='sparse',
    shuffle=False
)

# Define hyperparameters
LEARNING_RATE = 0.0001
EPOCHS = 15

# Function to create a model
def create_model(base_model):
    base_model.trainable = False  # Freeze the base model
    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(len(train_data.class_indices), activation='softmax')
    ])
    return model

# MobileNetV1 model
mobilenet_base = MobileNet(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3), include_top=False, weights='imagenet')
mobilenet_model = create_model(mobilenet_base)
mobilenet_model.compile(
    optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nTraining MobileNetV1...")
mobilenet_history = mobilenet_model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS
)

# ResNet50 model
resnet_base = ResNet50(input_shape=(IMG_HEIGHT, IMG_WIDTH, 3), include_top=False, weights='imagenet')
resnet_model = create_model(resnet_base)
resnet_model.compile(
    optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nTraining ResNet50...")
resnet_history = resnet_model.fit(
    train_data,
    validation_data=val_data,
    epochs=EPOCHS
)

# Evaluate and generate confusion matrix for MobileNetV1
def evaluate_and_plot(model, test_data, title):
    test_labels = test_data.classes
    predictions = model.predict(test_data)
    predicted_labels = np.argmax(predictions, axis=1)

    # Confusion Matrix
    cm = confusion_matrix(test_labels, predicted_labels)
    print(f"\nConfusion Matrix for {title}:")
    print(cm)

    # Classification Report
    print(f"\nClassification Report for {title}:")
    print(classification_report(test_labels, predicted_labels, target_names=test_data.class_indices.keys()))

    # Plot Confusion Matrix
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title(f"Confusion Matrix: {title}")
    plt.colorbar()
    tick_marks = np.arange(len(test_data.class_indices))
    plt.xticks(tick_marks, test_data.class_indices.keys(), rotation=45)
    plt.yticks(tick_marks, test_data.class_indices.keys())
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

print("\nEvaluating MobileNetV1...")
evaluate_and_plot(mobilenet_model, test_data, "MobileNetV1")

print("\nEvaluating ResNet50...")
evaluate_and_plot(resnet_model, test_data, "ResNet50")
```
